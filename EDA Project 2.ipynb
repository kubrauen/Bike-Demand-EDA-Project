# Startup-funding 

Startups have made significant contributions to various industries like e-commerce, online food delivery, transportation, education, technology etc over the past decade in India. Their new and innovative ideas always attracts investments from many sources. An analysis on the funding for Indian startups for the period 2015-2020 is done here.In the 'Startup-funding' data set; The basic steps of data preparation and cleaning, exploratory data analysis and visualization, and answering some interesting questions are covered here.

### Column Descriptions

Sr_No: Unique sequence number of each record.

Date: The date the investment was made.

Startup_Name: The name of the startup that received investment.

Industry_Vertical: The main industry in which the startup operates.

SubVertical: The sub-sector in which the startup operates.

City_Location: The city where the startup is located.

Investors_Name: Names of the people or organizations that invested.

InvestmentType: Type of investment made (e.g. Seed Funding, Series A, etc.).

Amount_In_Usd: The amount of the investment in US dollars.

Remarks: Additional information or notes about the investment.

Now by checking the values in 'Date' column against null values in 'DATE' column are checked. It is understood that 5 values are not converted due to mismatch of format that might have occurred from typo error. Now these values are also to be converted.

```python
# Creating a dictionary with indices of missing dates and thier orginal format.
date_dict = {
    'index' : [192,2571,2606,2775,2776,2831,3011,3029],
    'DATE' : ['05/07/2018','01/07/2015','10/07/2015','12/05/2015','12/05/2015','13/04/2015','15/01/2015','22/01/2015']
}
```

```python
import numpy as np

# NaN olarak değerlendirilecek liste
nan_list = ['Undisclosed', 'undisclosed', 'unknown', '\xc2\\xa0N/A']

def convert_numbers(number):
    try:
        # Eğer değer nan_list içinde yer alıyorsa NaN olarak ayarla
        if str(number).strip() in nan_list:
            return np.nan
        # Eğer değer belirli bir özel formattaysa belirli bir değere dönüştür
        elif number == '14,342,000+':
            return 14342000.0
        else:
            # Özel karakterleri temizle ve sayıya dönüştür
            cleaned_number = str(number).replace('\\xc2\\xa0', '').replace(',', '')
            return float(cleaned_number)
    except ValueError:
        # Sayıya dönüştürülemeyen durumlar için NaN döndür
        return np.nan
    except TypeError:
        # None tipi için de NaN döndür
        return np.nan

# Örnek kullanım
data = ['100', '200.5', '14,342,000+', 'Undisclosed', 'unknown', '300USD', '\xc2\\xa0N/A', None, 'invalid']

# Convert each item in the data list
converted_data = [convert_numbers(item) for item in data]

print(converted_data)
```

# Does Location also play a role, In determining the Growth of a Startup?

```python
most_common_locations = df['City_Location'].value_counts().head(15)
most_common_locations 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# İlk 15 lokasyonu al
most_common_locations = df['City_Location'].value_counts().head(15)

# Toplam yatırım sayısını hesapla
total_investments = df['City_Location'].value_counts().sum()

# Grafik boyutunu ve x ekseninin döndürülme açısını ayarla
plt.figure(figsize=(14,6))
plt.xticks(rotation=75)

# Grafik başlığını ve ekseni etiketlerini ayarla
plt.title('Location Role in Determining the Growth of a Startup ')
plt.xlabel('Location')
plt.ylabel('Count of Investment')

# Çubuk grafiği oluştur
barplot = sns.barplot(x=most_common_locations.index, y=most_common_locations).set(xlabel='Location', ylabel='Count of Investment')

# Yüzdelik dilimleri hesaplayarak barların üzerine ekle
for index, value in enumerate(most_common_locations):
    percentage = (value / total_investments) * 100
    plt.text(index, value, f'{percentage:.2f}%', ha='center', va='bottom')

plt.show()
```

It can be seen that most liked location for Indian startups is Bangalore, followed by Mumbai, New Delhi and Gurgaon

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Toplam yatırım miktarına göre ilk 15 lokasyonu hesapla
total_investment_by_location = df.groupby('City_Location')['Amount_Usd'].sum().sort_values(ascending=False).head(15)

# Toplam yatırım miktarını hesapla (yüzde hesaplamaları için)
total_investment = df['Amount_Usd'].sum()

# Grafik boyutunu ve x ekseninin döndürülme açısını ayarla
plt.figure(figsize=(14,6))
plt.xticks(rotation=75)

# Grafik başlığını ve ekseni etiketlerini ayarla
plt.title('Location Role in Determining the Growth of a Startup')
plt.xlabel('Location')
plt.ylabel('Total Investment Amount in USD')

# Çubuk grafiği oluştur
barplot = sns.barplot(x=total_investment_by_location.index, y=total_investment_by_location)

# Yüzdelik dilimleri hesaplayarak barların üzerine ekle
for index, value in enumerate(total_investment_by_location):
    percentage = (value / total_investment) * 100
    plt.text(index, value, f'{percentage:.2f}%', ha='center', va='bottom')

plt.show()
```

# Conclusions


We have observed many details of the data, some are stated below.

-Even though the average investment amount per transaction is 18.4 million USD, more than 75% of investments are below 10 million. It is observed that there are very few investments of more than 500 million USD are reason for the high average and most most common investment amounts are 1 million USD and 50000 USD.

-The most liked startup locations in India are 'Bangalore', 'Mumbai', 'New Delhi' and 'Gurgaon'.

-Over the years more startups have risen in Consumer Internet(roughly half in number) industry, Technology sector and e-Commerse. Together they make 85% of startups.

-Almost 90% of the investments are received as 'Seed Funding' or 'Private Equity'.

Among different industry verticals, e-Commerce industry has major share of investments (around 30%), the Consumer Internet industry has only close to 1/4 th share of investments. Also two industries with fewer number of startups, namely Finance and Transport have more than 10% of investment share each.

 -'Westbridge Capital' and 'Softbank' are major investors in Indian startups for the period.
There is an increase in investments in Indian startups from 2015 to 2020. Also in the third quarter (July - September), there considerable is spikes in investments.

-E-Commerse industry is consistent in attracting investment over the years and in 2017 & 2019 maximum funds were received.
